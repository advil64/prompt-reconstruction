```python
import os
import pandas as pd

# Specify the directory path
directory_path = '/Users/vigneshkrishnan/PoAI_Data'

# List all subdirectories in the directory
subdirs = [d for d in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, d))]

# Print the names of the subdirectories
# print("Subdirectories in the directory:")
# for subdir in subdirs:
#     print(subdir)

file_names = []
texts = []
transformers=[]

# Print the names of the subdirectories and their files
for subdir in subdirs:
    print(f"Subdirectory: {subdir}")
    
    # Construct the path to the subdirectory
    subdir_path = os.path.join(directory_path, subdir)
    
    # List all files in the subdirectory
    files = [f for f in os.listdir(subdir_path) if os.path.isfile(os.path.join(subdir_path, f))]
    
    # Print the names of the files in the subdirectory
    for file in files:
        file_path=subdir_path+"/"+file
        print(f" - {file_path}")
        print(f" - {file}")
        if os.path.isfile(file_path):
            # Read the text from the file
            with open(file_path, 'r', encoding='utf-8') as filee:
                text = filee.read()
                
            name_without_extension, _ = os.path.splitext(file)
            
            # Append the file name and text to the lists
            file_names.append(name_without_extension)
            texts.append(text)
            transformers.append(subdir)

# Create a DataFrame with the collected file names and texts
df1 = pd.DataFrame({
    'FileName': file_names,
    'Text': texts,
    'Transformer': transformers
})

# Print the DataFrame to verify
print(df1)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Example DataFrame
# df = pd.DataFrame({
#     'Text1': ['apple banana mango', 'coffee tea', 'bike car airplane'],
#     'Text2': ['banana orange', 'coffee milk tea', 'bike car train']
# })

# Function to find word differences between two texts
def find_word_differences(text1, text2):
    words1 = set(text1.split())
    words2 = set(text2.split())
    return ' '.join(words1.symmetric_difference(words2))

# Apply the function to each row and store the results in a new column
merged_df['Differences'] = merged_df.apply(lambda row: find_word_differences(row['Text'], row['Paragraph']), axis=1)

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'Differences' column to create TF-IDF embeddings
tfidf_embeddings = tfidf_vectorizer.fit_transform(merged_df['Differences'])

# Print TF-IDF embeddings
print(tfidf_embeddings)

# If you want to view the embeddings as a dense matrix (for small datasets)
print(tfidf_embeddings.toarray())

# To get feature names to understand which column corresponds to which word
print(tfidf_vectorizer.get_feature_names_out())


merged_df[['Text','Paragraph','Differences']]

merged_df.to_csv('word-diffs.csv')
```
