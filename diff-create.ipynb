```python
import os
import pandas as pd

# Specify the directory containing your files
directory_path = './new-data'

# Specify the string that the filenames should start with
filename_starts_with = 'Original_Text'

# Initialize lists to store filenames and their content
filenames_og = []
contents_og = []
labels_og=[]

filenames_new = []
contents_new = []
labels_new = []

for directory in os.listdir(directory_path):
    print(directory)
    new_path = os.path.join(directory_path, directory)
    
    # Ensure that we're iterating over directories
    if not os.path.isdir(new_path):
        continue

    for filename in os.listdir(new_path):
        print(filename)
        # Check if the filename starts with the specified string
        if filename.startswith(filename_starts_with):
            
            name_without_extension, _ = os.path.splitext(filename)
            # Correctly construct the full path to the file
            file_path = os.path.join(new_path, filename)  # Use new_path instead of directory_path
            
            print(file_path)
            
            # Read the content of the file
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                
                # Append the filename and content to the lists
                filenames_og.append(name_without_extension.replace('Original_Text_', ''))
                contents_og.append(content)
                labels_og.append(directory)
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")
                
        else:
            name_without_extension, _ = os.path.splitext(filename)
            # Correctly construct the full path to the file
            file_path = os.path.join(new_path, filename)  # Use new_path instead of directory_path
            
            print(file_path)
            
            # Read the content of the file
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                
                # Append the filename and content to the lists
                filenames_new.append(name_without_extension)
                contents_new.append(content)
                labels_new.append(directory)
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")

# Create a DataFrame from the filenames and contents
df_og = pd.DataFrame({
    'Filename_OG': filenames_og,
    'Content_OG': contents_og,
    'Labels_OG': labels_og
})

df_new = pd.DataFrame({
    'Filename_New': filenames_new,
    'Content_New': contents_new,
    'Labels_New': labels_new
})

print(df_og)

print(df_new)

# # Specify the path to save the CSV file
# csv_file_path = '/path/to/your/filename.csv'
# 
# # Save the DataFrame to a CSV file
# df.to_csv(csv_file_path, index=False)
```

```python
# Merge using different column names
merged_df = pd.merge(df_og, df_new, left_on=['Filename_OG', 'Labels_OG'], right_on=['Filename_New', 'Labels_New'], how='inner')

print(merged_df)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Example DataFrame
# df = pd.DataFrame({
#     'Text1': ['apple banana mango', 'coffee tea', 'bike car airplane'],
#     'Text2': ['banana orange', 'coffee milk tea', 'bike car train']
# })

# Function to find word differences between two texts
def find_word_differences(text1, text2):
    words1 = set(text1.split())
    words2 = set(text2.split())
    return ' '.join(words1.symmetric_difference(words2))

def find_word_differences_new(text1, text2):
    words1 = set(text1.split())
    words2 = set(text2.split())
    # Use set difference to find words in words2 not in words1
    return ' '.join(words2 - words1)


# Apply the function to each row and store the results in a new column
merged_df['Differences'] = merged_df.apply(lambda row: find_word_differences_new(row['Content_OG'], row['Content_New']), axis=1)

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'Differences' column to create TF-IDF embeddings
tfidf_embeddings = tfidf_vectorizer.fit_transform(merged_df['Differences'])

# Print TF-IDF embeddings
print(tfidf_embeddings)

# If you want to view the embeddings as a dense matrix (for small datasets)
print(tfidf_embeddings.toarray())

# To get feature names to understand which column corresponds to which word
print(tfidf_vectorizer.get_feature_names_out())


print(merged_df[['Content_OG','Content_New','Differences']])

merged_df.to_csv('new-word-diffs2.csv')
```

```python
import os
import pandas as pd

data=merged_df

# Define the root directory name
root_dir = "diff-files-new"

# Ensure the root directory exists
os.makedirs(root_dir, exist_ok=True)

# Iterate over the DataFrame to create directories and files
for index, row in data.drop_duplicates('Labels_OG').iterrows():
    # Directory path for the current label
    label_dir_path = os.path.join(root_dir, row['Labels_OG'])
    
    # Ensure the directory exists
    os.makedirs(label_dir_path, exist_ok=True)
    
    # Filter rows with the current label to create files
    for _, file_row in data[data['Labels_OG'] == row['Labels_OG']].iterrows():
        file_path = os.path.join(label_dir_path, file_row['Filename_OG'])
        
        # Create and write to the file
        with open(file_path, 'w') as file:
            file.write(file_row['Differences'])

print("Directories and files have been created successfully.")
```
